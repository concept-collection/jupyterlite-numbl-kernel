# Control flow and functions

Branches, loops, anonymous functions, and named functions kept in `.m`
files alongside the notebook.

## if / elseif / else

```numbl
x = 7;
if mod(x, 2) == 0
    disp('even')
elseif x > 5
    disp('odd and > 5')
else
    disp('odd and <= 5')
end
```

## for and while loops

```numbl
total = 0;
for k = 1:10
    total = total + k;
end
total
```

```numbl
n = 0; p = 1;
while p < 1000
    n = n + 1;
    p = p * 2;
end
fprintf('2^%d = %d is the first power of two >= 1000\n', n, p);
```

## switch

```numbl
grade = 'B';
switch grade
    case 'A'
        disp('excellent')
    case {'B', 'C'}
        disp('good')
    otherwise
        disp('try again')
end
```

## Anonymous functions, arrayfun, and cellfun

```numbl
sq = @(x) x.^2;
sq(1:5)
arrayfun(@(k) k^2, 1:5)
cellfun(@numel, {'a', 'bb', 'ccc'})
```
