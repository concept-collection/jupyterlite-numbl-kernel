# Linear algebra

numbl uses LAPACK for its factorizations (native on the command line,
WebAssembly in the browser).

## Solving a linear system

```numbl
A = [4 -2 1; -2 4 -2; 1 -2 3];
b = [1; 2; 3];
x = A \ b
residual = norm(A * x - b)
```

## Determinant, rank, trace, inverse

```numbl
A = [4 -2 1; -2 4 -2; 1 -2 3];
det(A)
rank(A)
trace(A)
inv(A)
```

## Eigenvalues and eigenvectors

```numbl
A = [4 -2 1; -2 4 -2; 1 -2 3];
[V, D] = eig(A);
eigenvalues = diag(D)'
```

## Singular values and QR

```numbl
A = [4 -2 1; -2 4 -2; 1 -2 3];
s = svd(A)
[Q, R] = qr(A);
orthogonality_error = norm(Q' * Q - eye(3))
```

## Vector and matrix norms

```numbl
A = [4 -2 1; -2 4 -2; 1 -2 3];
norm([3 4])
norm([3 4], Inf)
norm(A, 'fro')
```
