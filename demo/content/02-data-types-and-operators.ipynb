# Data types and operators

A tour of numbl's scalar types and operators.

## Numbers and arithmetic

```numbl
x = 3.5;
y = 2;
[x + y, x - y, x * y, x / y]
x ^ y
mod(7, 3)
```

## Logical values and comparisons

```numbl
v = [1 2 3 4 5];
v > 2
v == 3
any(v > 4)
all(v > 0)
```

```numbl
t = true; f = false;
[t && f, t || f, xor(t, f), ~t]
```

## Complex numbers

```numbl
z = 3 + 4i;
abs(z)
angle(z)
[real(z), imag(z)]
conj(z)
```

## Constants, special values, and rounding

```numbl
[pi, exp(1), eps]
[1/0, -1/0, 0/0]
[isnan(0/0), isinf(1/0)]
```

```numbl
[round(2.5), floor(-2.3), ceil(2.1), fix(-2.7), sign(-3)]
```
