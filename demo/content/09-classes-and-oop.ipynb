# Classes and object-oriented programming

numbl supports MATLAB's `classdef` object model: value and handle classes,
constructors, methods, inheritance, abstract methods, operator overloading,
and static methods. Classes are defined in `.m` files next to the notebook
(you can open and edit them in the file browser).

## A value class with operator overloading

[Vec2.m](./Vec2.m) overloads `+`, `-`, and `*` and defines `norm`.

```numbl
a = Vec2(3, 4);
b = Vec2(1, 2);
fprintf('a + b = %s\n', char(a + b));
fprintf('a - b = %s\n', char(a - b));
fprintf('3 * a = %s\n', char(3 * a));
fprintf('norm(a) = %g\n', norm(a));
```

## A handle class has reference semantics

[Account.m](./Account.m) is a `handle` class, so assigning it does not copy:
both names refer to the same object.

```numbl
acct = Account(100);
deposit(acct, 50);
alias = acct;
deposit(alias, 25);
acct.Balance     % 175: updates are seen through both names
```

## Inheritance, abstract methods, and polymorphism

[Shape.m](./Shape.m) declares an abstract `area()` and a concrete
`describe()`. [Circle.m](./Circle.m) and [Square.m](./Square.m) implement
`area()`.

```numbl
shapes = {Circle(2), Square(3)};
areas = cellfun(@(s) s.area(), shapes)
cellfun(@(s) isa(s, 'Shape'), shapes)
```

```numbl
for i = 1:numel(shapes)
    shapes{i}.describe();
end
```

## Static methods and methods across files

[MathX.m](./MathX.m) has a static method (called on the class, not an
instance). [@Poly/](./@Poly/Poly.m) is a class whose methods live in
separate files in its `@Poly` folder.

```numbl
MathX.hypot3(3, 4, 12)
p = Poly([1 -3 2]);     % x^2 - 3x + 2
p.evalAt(5)             % method from @Poly/evalAt.m
roots(p.coeffs)'
```
