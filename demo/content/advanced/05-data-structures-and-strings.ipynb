# Data structures and strings

Cell arrays, structs, dictionaries, and text handling.

## Cell arrays hold mixed types

```numbl
c = {42, 'hello', [1 2 3]};
c{2}
class(c{3})
numel(c)
```

## Structs

```numbl
p.name = 'ada';
p.year = 1815;
p.tags = {'math', 'cs'};
fieldnames(p)
p.tags{1}
```

## Struct arrays

```numbl
people(1).name = 'A'; people(1).age = 30;
people(2).name = 'B'; people(2).age = 25;
ages = [people.age]
names = {people.name}
```

## Key/value maps with containers.Map

```numbl
prices = containers.Map({'apple', 'banana', 'cherry'}, {3, 5, 7});
prices('banana')
keys(prices)
```

## Formatting and parsing text

```numbl
s = sprintf('%d items at $%.2f = $%.2f', 3, 1.50, 4.50)
num2str(pi, 8)
str2double('3.14')
```

## String functions

```numbl
strsplit('a,b,c', ',')
strjoin({'x', 'y', 'z'}, '-')
upper('numbl')
strrep('banana', 'a', 'o')
```

## String arrays and regular expressions

```numbl
names = ["Ada" "Alan" "Grace"];
names(2)
contains(names, "A")
regexp('id=42; val=7', '\d+', 'match')
```
