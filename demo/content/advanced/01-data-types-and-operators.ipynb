# Data types and operators

A tour of numbl's scalar types and operators. numbl uses MATLAB syntax, so
everything here matches what you would type in MATLAB. Run each cell with
**Shift+Enter**; unsuppressed expressions print their value.

## Numbers and arithmetic

```numbl
x = 3.5;
y = 2;
[x + y, x - y, x * y, x / y]
x ^ y
mod(7, 3)
```

## Integer types saturate

```numbl
a = int8(200)     % clamps to 127
b = uint8(-5)     % clamps to 0
class(a)
double(a) + 0.5
```

## Logical values and comparisons

```numbl
v = [1 2 3 4 5];
v > 2
v == 3
any(v > 4)
all(v > 0)
```

```numbl
t = true; f = false;
[t && f, t || f, xor(t, f), ~t]
```

## Complex numbers

```numbl
z = 3 + 4i;
abs(z)
angle(z)
[real(z), imag(z)]
conj(z)
```

## Special values and rounding

```numbl
[1/0, -1/0, 0/0]
[isnan(0/0), isinf(1/0)]
[pi, eps, Inf]
```

```numbl
[round(2.5), floor(-2.3), ceil(2.1), fix(-2.7), sign(-3)]
```
