# Numerical methods

Polynomials, interpolation, the FFT, statistics, and quadrature.

## Polynomials: fit, evaluate, roots

```numbl
x = 0:0.5:5;
y = 2 * x.^2 - 3 * x + 1;
p = polyfit(x, y, 2)
polyval(p, 10)
roots([1 -3 2])'
```

## Interpolation

```numbl
xs = 0:10;
ys = xs.^2;
interp1(xs, ys, 3.5)
interp1(xs, ys, [1.5 5.5 8.5])
```

## The FFT

```numbl
t = 0:0.01:1 - 0.01;
sig = sin(2 * pi * 5 * t) + 0.5 * sin(2 * pi * 12 * t);
F = abs(fft(sig));
[~, idx] = max(F(1:50));
fprintf('dominant frequency near %d Hz\n', idx - 1);
```

## Basic statistics

```numbl
data = [4 8 15 16 23 42];
[mean(data), median(data), std(data)]
[min(data), max(data)]
```

## Sorting, differences, cumulative sums

```numbl
v = [3 1 4 1 5 9 2 6];
sort(v)
unique(v)
cumsum(v)
diff(v)
```
